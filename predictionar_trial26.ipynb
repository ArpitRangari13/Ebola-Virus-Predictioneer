```python
# First, we need to load Covid-19 dataset from GitHub Directly. You may load it from your computer but I found load it directly from the John Hopkins repository give a lot of advantages.
# 1) You don't have to load it everyday. 2) You will save your hard drive space.
import pandas as pd
url='https://docs.google.com/spreadsheets/d/1lMcoJVzBb1yEBSZNxjMKjNELoc8o5FbP/edit?usp=sharing&ouid=102292434359970313631&rtpof=true&sd=true'
df = pd.read_csv(url)
```

```python
# After loading, let's take a look at the data.
df.head()
```

```python
# Inspect the original DataFrame
print("Original DataFrame:")
print(df.head())
print("Columns in the original DataFrame:")
print(df.columns.tolist())

# Check for whitespace and adjust column names
df.columns = df.columns.str.strip()  # Remove whitespace

# Print adjusted column names
print("Adjusted Columns:")
print(df.columns.tolist())

# Drop unnecessary columns
try:
    ddf = df.drop(columns=['Province/State', 'Lat', 'Long'])
    print("\nModified DataFrame after dropping columns:")
    print(ddf.head())
except KeyError as e:
    print(f"KeyError: {e}. Please check the column names.")
```

```python
# Group by 'Latitude' and 'Longitude' and sum the values
gddf = ddf.groupby(['Lat', 'Long_']).sum()

# Reset the index to convert the grouped DataFrame back to a regular DataFrame
gddf = gddf.reset_index()

# Inspect the grouped DataFrame
print("\nGrouped DataFrame by Latitude and Longitude:")
print(gddf.head())
```

```python
# If you plan to predict trends for your country, find your country name here.
# The country name has to be identical to the listed ones for the program to work.
for x in gddf.T.columns:
  print(x)
```

```python
# Here if you want to analyze your country's data, simply replace 'US' with your counntry name below.
# You need to keep the single quotes.
y = gddf.T['US']
```

```python
# Ok. Let's see what your country's death toll trend looks like graphically.
y.plot()
```

```python
# Since we are predicting daily death tolls, we need to adjust data from accumulated death tolls to daily death tolls.
for i in range(y.size-1,1,-1):
  y.iloc[i] = y.iloc[i]-y.iloc[i-1]
```

```python
y.plot()
```

```python
# Before machine learning can do for you, you need to do data preprocessing for training data in x dimension
# Also needed is normalize xtrain and let all values in [0.0, 1.0).
xtrain = list(range(y.size))
xtrain = [x/y.size for x in xtrain]
```

```python
# Do data preprocessing for training data in y dimension as well.
# Note that y is a Pandas Series but ytrain is a list. They are differnt objects in Python.
ymax = y.max()
ytrain = y.div(ymax+1)
ytrain = ytrain.to_list()
```

```python
# It's time to get Tensorflow library in and create a model.
import tensorflow as tf
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=[1]),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(1)
])
```

```python
# Take a look at the model you have just created.
model.summary()
```

```python
# Before taining, you need to configure the model for training.
model.compile(optimizer='Adam',
              loss='mse',
              metrics=['mae','mse'])
```

```python
# For your model to be able to predict something, your need to train your model using training data.
history = model.fit(xtrain, ytrain, epochs=500)
```

```python
# Convert training history to dataframe for viewing and graphing.
hist = pd.DataFrame(history.history)
hist.head()
```

```python
# Training errors weith respect to ecochs.
hist.plot()
```

```python
# Now, we can predict today's death toll (June 4, 2020).
# Note that we normalize x to [0, 1). So today's x value is 1.0.
xtest=1.0
predicted_death_today = (1+ymax)*model.predict([xtest])
print(int(predicted_death_today[0]))
```

```python
# We may graphic all predcited values along with x values.
# First run model predictions for all x values.
predicted = (1*ymax)*model.predict(xtrain)
```

```python
# Convert numpy ndarray to pandas series to use Pandas plot function.
yp=pd.Series(predicted[:,0])
```

```python
# Plot original and predicted death tolls together to compare the leraning curve to the orignal data.
y.plot(legend='true');yp.plot()
```
